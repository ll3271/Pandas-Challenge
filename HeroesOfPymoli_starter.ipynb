## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
No_Of_Male = purchase_data.loc[(purchase_data['Gender']== 'Male')].count()

No_Of_Female = purchase_data.loc[(purchase_data['Gender']== 'Female')].count()

No_Of_Others = purchase_data.loc[(purchase_data['Gender']== 'Other / Non-Disclosed')].count()

Percentage_Of_Female = No_Of_Female['Gender'] / purchase_data['Gender'].count()

Percentage_Of_Male = No_Of_Male['Gender'] / purchase_data['Gender'].count()

Percentage_Of_Others = No_Of_Others['Gender'] / purchase_data['Gender'].count()
```

```python
Gender_Of_Players = {'Gender': ['Male','Female','Other / Non-Disclosed'],
        'Total Count': [No_Of_Male['Gender'],No_Of_Female['Gender'],No_Of_Others['Gender']],
       'Percentage Of Players': [Percentage_Of_Male,Percentage_Of_Female,Percentage_Of_Others]}
Gender_Of_Players_df = pd.DataFrame(Gender_Of_Players,columns= ['Gender', 'Total Count','Percentage Of Players'])
Gender_Of_Players_df['Percentage Of Players'] = Gender_Of_Players_df['Percentage Of Players'] * 100
Gender_Of_Players_df['Percentage Of Players'] = Gender_Of_Players_df['Percentage Of Players'].map('{:,.2f}%'.format)
Gender_Of_Players_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9.9,13.9,18.9,23.9,28.9,33.9,38.9,99.9]
group_names = ["<10",'10-14','15-19','20-24','25-29','30-34','35-39','40+']
purchase_data['Age Ranges'] = pd.cut(purchase_data['Age'],bins, labels=group_names)
Age_demo = purchase_data.groupby(['Age Ranges']).agg ({'Purchase ID': 'count'})
Age_demo['Total Count'] = Age_demo['Purchase ID']
Age_demo['Percentage Of Players'] = Age_demo['Purchase ID'] / Age_demo['Purchase ID'].count()
Age_demo['Percentage Of Players'] = Age_demo['Percentage Of Players'].map('{:,.2f}%'.format)
del Age_demo['Purchase ID']
Age_demo
```
